```python
# prompt: groupby personid RR	PR	QRS	QT	QTc	P_wave_axis	QRS_axis	T_wave_axis and onehot encode labname in seperate column such that 1:potassium, 2: magnesium and 3:calcium also keep their lab value in their seperate column sch that for 1 we have its lab value in its column 1: potassium_lab_value and respectively for other minerals
temp_df = df
# Create dummy columns for one-hot encoding and lab values
temp_df['potassium'] = 0
temp_df['magnesium'] = 0
temp_df['calcium'] = 0
temp_df['potassium_lab_value'] = 0.0
temp_df['magnesium_lab_value'] = 0.0
temp_df['calcium_lab_value'] = 0.0

# One-hot encode labname and populate lab values
for index, row in temp_df.iterrows():
    if row['labname'] == 1:
        temp_df.loc[index, 'potassium'] = 1
        temp_df.loc[index, 'potassium_lab_value'] = row['labvalue']
    elif row['labname'] == 2:
        temp_df.loc[index, 'magnesium'] = 1
        temp_df.loc[index, 'magnesium_lab_value'] = row['labvalue']
    elif row['labname'] == 3:
        temp_df.loc[index, 'calcium'] = 1
        temp_df.loc[index, 'calcium_lab_value'] = row['labvalue']

# Drop the original 'labname' and 'labvalue' columns
temp_df = temp_df.drop(['labname', 'labvalue'], axis=1)

# Group by 'personid' and aggregate other columns
grouped_df = temp_df.groupby(['personid','RR','PR','QRS','QT','QTc','P_wave_axis','QRS_axis','T_wave_axis']).agg({
    'potassium': 'max',  # Use max since it's a one-hot encoding
    'magnesium': 'max',
    'calcium': 'max',
    'potassium_lab_value': 'max',
    'magnesium_lab_value': 'max',
    'calcium_lab_value': 'max'
}).reset_index()

grouped_df
```

```python

def get_stats(labname: int, low: float, high: float):
  # Define the labname you want to check and its normal range
  labname_to_check = labname  # Change this to the labname you want to check
  normal_range_min = low
  normal_range_max = high
  switch_labname = {
    1: "potassium",
    2: "magnesium",
    3: "calcium"
  }
  switch_labvalue = {
    1: "potassium_lab_value",
    2: "magnesium_lab_value",
    3: "calcium_lab_value"
  }
  # Filter rows where labname is the one we're interested in
  filtered_data = data[data[switch_labname[labname]] == 1]

  # Count total instances of this labname
  total_instances = len(filtered_data)

  # Count instances where labvalue is within the normal range
  normal_instances = filtered_data[(filtered_data[switch_labvalue[labname]] >= normal_range_min) & (filtered_data[switch_labvalue[labname]] <= normal_range_max)]

  # Calculate how many are within the normal range
  normal_count = len(normal_instances)

  # Print results
  print(f"Total instances of labname {labname_to_check}: {total_instances}")
  print(f"Instances within normal range ({normal_range_min} to {normal_range_max}): {normal_count}")
  print(f"Percentage of normal values: {normal_count / total_instances * 100:.2f}%")
  print()

ranges = [(1, 3.5,5),(2, 8.5,10.5),(3, 1.5,2.5)]
for i in ranges:
  get_stats(*i)
```
