```python


# Define the labname you want to check and its normal range
labname_to_check = 3  # Change this to the labname you want to check
normal_range_min = 1.5
normal_range_max = 2.5

# Filter rows where labname is the one we're interested in
filtered_data = data[data['labname'] == labname_to_check]

# Count total instances of this labname
total_instances = len(filtered_data)

# Count instances where labvalue is within the normal range
normal_instances = filtered_data[(filtered_data['labvalue'] >= normal_range_min) & (filtered_data['labvalue'] <= normal_range_max)]

# Calculate how many are within the normal range
normal_count = len(normal_instances)

# Print results
print(f"Total instances of labname {labname_to_check}: {total_instances}")
print(f"Instances within normal range ({normal_range_min} to {normal_range_max}): {normal_count}")
print(f"Percentage of normal values: {normal_count / total_instances * 100:.2f}%")
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, TensorBoard
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import datetime
```

```python
data
```

```python
# Separate features and targets
X = data[['RR', 'PR', 'QRS', 'QT', 'QTc', 'P_wave_axis', 'QRS_axis', 'T_wave_axis']].values  # Replace with actual feature columns
y_labname = data['labname'].values  # Classification target
y_labvalue = data['labvalue'].values  # Regression target

```

```python
# One-hot encode the classification target
y_labname = data['labname'].values  
```

```python
y_labname = y_labname - 1  # Adjust if necessary to ensure indices start from 0 (0, 1, 2 for classes)
num_classes = len(np.unique(y_labname))  # Should be 3 in this case
y_labname = tf.keras.utils.to_categorical(y_labname, num_classes=num_classes)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Reshape X for CNN input
X = X.reshape(X.shape[0], X.shape[1], 1)
```

```python


# Split the data into training and testing sets
X_train, X_test, y_train_labname, y_test_labname, y_train_labvalue, y_test_labvalue = train_test_split(
    X, y_labname, y_labvalue, test_size=0.2, random_state=42
)
```

```python
inputs = Input(shape=(X.shape[1], 1))


x = LSTM(64, return_sequences=True)(inputs)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = LSTM(128, return_sequences=False)(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = Dense(128, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.5)(x)

x = Dense(64, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.5)(x)

x = Dense(256, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.5)(x)

x = Dense(256, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)


output_class = Dense(num_classes, activation='softmax', name='classification_output')(x)


output_regression = Dense(1, activation='linear', name='regression_output')(x)


model = Model(inputs=inputs, outputs=[output_class, output_regression])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),  
    loss={'classification_output': 'categorical_crossentropy', 'regression_output': 'mse'},
    metrics={'classification_output': 'accuracy', 'regression_output': 'mae'}
)
model.summary()
```

```python
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
```

```python
callbacks = [
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6),
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    TensorBoard(log_dir=log_dir, histogram_freq=1)
]
```

```python
model.summary()
```

```python
model.fit(
    X_train,
    {'classification_output': y_train_labname, 'regression_output': y_train_labvalue},
    epochs=50,
    batch_size=32,
    validation_data=(X_test, {'classification_output': y_test_labname, 'regression_output': y_test_labvalue}),
    callbacks=callbacks
)
```

```python
# Evaluate the model
loss, class_loss, reg_loss, class_acc, reg_mae = model.evaluate(
    X_test,
    {'classification_output': y_test_labname, 'regression_output': y_test_labvalue}
)

print(f'Test Loss: {loss}, Classification Accuracy: {class_acc}, Regression MAE: {reg_mae}')
```
