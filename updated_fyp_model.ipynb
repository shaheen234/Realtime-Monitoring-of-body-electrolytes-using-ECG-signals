```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, TensorBoard
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import datetime
```

```python
# Separate features and targets
X = data[['RR', 'PR', 'QRS', 'QT', 'QTc', 'P_wave_axis', 'QRS_axis', 'T_wave_axis']].values  # Replace with actual feature columns
y_labname = data['labname'].values  # Classification target
y_labvalue = data['labvalue'].values  # Regression target

```

```python
# One-hot encode the classification target
y_labname = data['labname'].values  
```

```python
y_labname = y_labname - 1  # Adjust if necessary to ensure indices start from 0 (0, 1, 2 for classes)
num_classes = len(np.unique(y_labname))  # Should be 3 in this case
y_labname = tf.keras.utils.to_categorical(y_labname, num_classes=num_classes)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Reshape X for CNN input
X = X.reshape(X.shape[0], X.shape[1], 1)
```

```python


# Split the data into training and testing sets
X_train, X_test, y_train_labname, y_test_labname, y_train_labvalue, y_test_labvalue = train_test_split(
    X, y_labname, y_labvalue, test_size=0.2, random_state=42
)
```

```python
import numpy as np
import pandas as pd
from sklearn.utils import class_weight
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout, BatchNormalization, Bidirectional
from tensorflow.keras.optimizers import Adam

# Assuming X_reshaped, y_labname, y_labvalue are your preprocessed inputs
# Reshape X to 3D (samples, timesteps, features) as required for LSTM
X_reshaped = X.reshape((X.shape[0], X.shape[1], 1))

# Define the number of classes for multilabel classification (use sigmoid output)
num_classes = y_labname.shape[1]

# Calculate class weights for each label (since it's multilabel, you might need weights for each label)
# You can skip this step if weights aren't necessary for the multilabel case
class_weights = class_weight.compute_class_weight('balanced',
                                                  classes=np.unique(np.argmax(y_labname, axis=1)),
                                                  y=np.argmax(y_labname, axis=1))
class_weights_dict = dict(enumerate(class_weights))

# LSTM model architecture
inputs = Input(shape=(X_reshaped.shape[1], 1))

# Bidirectional LSTM layers
x = Bidirectional(LSTM(128, return_sequences=True))(inputs)
x = BatchNormalization()(x)
x = Dropout(0.2)(x)

x = Bidirectional(LSTM(256, return_sequences=False))(x)
x = BatchNormalization()(x)
x = Dropout(0.2)(x)

# Dense layers for further processing
x = Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001))(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001))(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

# Output for multilabel classification (sigmoid activation for multilabel classification)
output_class = Dense(num_classes, activation='sigmoid', name='classification_output')(x)

# Output for regression (labvalue)
output_regression = Dense(1, activation='linear', name='regression_output')(x)

# Create the model
model = Model(inputs=inputs, outputs=[output_class, output_regression])

# Compile the model with binary cross-entropy for multilabel classification
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss={'classification_output': 'binary_crossentropy', 'regression_output': 'mse'},
    metrics={'classification_output': ['binary_accuracy'], 'regression_output': ['mae']}
)

# Model summary
model.summary()

# Train the model (with or without sample weights for classification)
```

```python
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
```

```python
callbacks = [
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6),
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    TensorBoard(log_dir=log_dir, histogram_freq=1)
]
```

```python
history = model.fit(
    X_reshaped,
    {'classification_output': y_labname, 'regression_output': y_labvalue},
    epochs=50,
    batch_size=128,
    validation_split=0.2
)
```

```python
# Evaluate the model
loss, class_loss, reg_loss, class_acc, reg_mae = model.evaluate(
    X_test,
    {'classification_output': y_test_labname, 'regression_output': y_test_labvalue}
)

print(f'Test Loss: {loss}, Classification Accuracy: {class_acc}, Regression MAE: {reg_mae}')
```
